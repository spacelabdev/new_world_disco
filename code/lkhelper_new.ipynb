```python
import pandas as pd
import lightkurve as lk
```

```python
valid_authors = ["Kepler","K2", "SPOC","TESS-SPOC","QLP","TASOC","PATHOS","CDIPS","K2SFF","EVEREST","TESScut","GSFC-ELEANOR-LITE"]
```

```python
tess_data_master = pd.read_csv('tess_data.csv')
tess_data_master.head()
```

```python
tess_id_list = tess_data_master['TIC ID']
```

```python
id_string = f'TIC {231663901}'

q = lk.search_lightcurve(id_string)

list_of_columns = list(q.table.columns)
```

```python
whole_df = ''

#for t in tess_id_list[:70]
for i in range(10):

    print(f'At step {i}') #Counter for tracking where the current iteration of the loop
    t = tess_id_list[i]

    id_string = f'TIC {t}'

    q = lk.search_lightcurve(id_string)

    # Stores all of the authors from different observations
    # For a given lightcurve

    columns_master_list = list(q.table.columns)
    columns_to_add = []
    for col in columns_master_list:
        c = q.table[col]

        c = pd.DataFrame(c)

        columns_to_add.append(c)

    temp_df = pd.concat(columns_to_add, axis=1)
    temp_df.columns = columns_master_list

    if t == tess_id_list[0]:
        whole_df = temp_df
    else:
        whole_df = pd.concat([whole_df, temp_df], axis=0, ignore_index=True)


whole_df.head()
```
